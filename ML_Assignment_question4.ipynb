```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import os

wineQ = pd.read_csv("/content/drive/MyDrive/College stuff/5th Semester/CS550 Machine Learning/Assignments/Assignment1/WineQT.csv")
print(wineQ.shape)

wineQ.head()

wineQ['quality'] = wineQ['quality'].apply({(lambda x:1 if x>6 else 0)})
wineQ
```

```python

#to normalize the values to between 0 to 1 we use MIn-Max Scaling
sc = MinMaxScaler()

#columns to not apply min max scaling on
columns_excl = ['Quality', 'Id']

#to check for data types within the columns
#wineQ.dtypes

columns_incl = wineQ.columns.difference(columns_excl)
#fit the df into the MinMaxScaler obj created
sc.fit(wineQ[columns_incl])
#applying transform
wineQ_scaled = wineQ.copy()
wineQ_scaled[columns_incl] = sc.transform(wineQ[columns_incl])

wineQ_scaled.to_csv(os.path.join('/content/drive/MyDrive/College stuff/5th Semester/CS550 Machine Learning/Assignments/Assignment1','WineQ_scaled.csv'), index=False)
print(wineQ_scaled)
```
